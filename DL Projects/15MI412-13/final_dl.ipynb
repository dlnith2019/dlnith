```
import numpy as np
from keras.models import Sequential
from keras.layers import SimpleRNN, LSTM
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```
X= [[[0+j] for j in np.arange(i-1,i,0.1)] for i in np.arange(1,32)]
print(*X, sep="\n")

Y= [(0+i) for i in range(1,32)]
print (Y)

```

```
Y1= np.reshape(Y,(31,1))
X1= np.asarray(X)
```

```
print(X1)
```

```
print(Y1)
```

```
X1 = np.array(X1, dtype="float32")
Y1 = np.array(Y1, dtype="float32")
```

```
X1.shape
```

```
Y1.shape
```

```
X_train, X_test, y_train, y_test = train_test_split(X1,Y1,test_size=0.2, 
                                                random_state=5)
```

```
from keras.layers import SimpleRNN,LSTM,Flatten, Dense
model = Sequential()
#model.add(LSTM((2),input_shape=(10,1),return_sequences=True))
#model.add(LSTM((3),input_shape=(10,1),return_sequences=True))

#model.add(SimpleRNN(5, input_shape=(31,1)))
#model.add(Dense((31,1), activation='relu'))
#model.add(Dense(3,activation='relu'))
#model.compile(optimizer='adam',loss='mae',metrics=['acc'])
#model.fit(X, Y, epochs=500, batch_size=64, verbose=2)

model.add(LSTM(200, activation='relu', return_sequences=True, input_shape=(10,1)))
model.add(LSTM(100, activation='relu', return_sequences=True))
model.add(LSTM(50, activation='relu', return_sequences=True))
model.add(LSTM(25, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
print(model.summary())
```

```
hist = model.fit(X_train, y_train, epochs=1000, batch_size=64, validation_data=(X_test, y_test))
```

```
y_predict = model.predict(X_test)
```
