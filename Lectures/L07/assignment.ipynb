```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
from keras.utils import np_utils
from keras.models import load_model
import numpy as np

x=[]
y=[]
for i in range(0,100):
    if(i<50):
        y.append(0)
    else:
        y.append(1)
    x.append(i)
#print(x)
#print(y)
x=np.array(x)
y=np.array(y)
x=x.astype(np.float)
y=y.astype(np.float)
x/=99
#print(x)

```

# LOSS FUNCTIONS:

Some of the loss functions are implemented and visualized below. Refer to https://keras.io/losses/ for more types of loss functions

#  Visualizing mean squared loss function

```python
import matplotlib.pyplot as plt

model=Sequential()
model.add(Dense(20,input_shape=(1,)))
model.add(Dense(1,activation="sigmoid"))
#model.summary()
model.compile(loss="mse",optimizer=SGD(lr=1),metrics=['accuracy'])



history = model.fit(x, y, validation_split=0.25, epochs=50, batch_size=16, verbose=0)

# Plot training & validation accuracy values
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train','Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train','Test'], loc='upper left')
plt.show()
```

# REGULARIZERS

Below we visualize three different regularizer. For Lasso regularizer we will furthur visualize the loss for different values. Visit https://keras.io/regularizers/ for more information on regualrizers and how to define a new one.

# Visualizing Ridge regularizer

```python
import matplotlib.pyplot as plt
from keras import regularizers
model=Sequential()
model.add(Dense(20,input_shape=(1,)))
model.add(Dense(1,activation="sigmoid",kernel_regularizer=regularizers.l2(0.01)))
#model.summary()
model.compile(loss="mse",optimizer=SGD(lr=1),metrics=['accuracy'])



history = model.fit(x, y, validation_split=0.25, epochs=50, batch_size=16, verbose=0)

# Plot training & validation accuracy values
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train','Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train','Test'], loc='upper left')
plt.show()
```
