We first define two sequences as follows :<br/>

Sequence 1 : <br/>
X1[i]= X1[i-1]+i*sin(i) for i =2,3,.....101<br/>
X1[1]=1<br/>
Sequence 2 : X2[i]= X2[i-1]+i*sin(5*i) for i =2,3,.....101<br/>
 X1[1]=1<br/>
At a given time we feed in,<br/>
 X1[i],X1[i+1], X2[i], X2[i+1] and the output is given by <br/>
          Y[i]= (X1[i] -  X2[i])+(X1[i+1]-X2[i+1]) for i=1,2.....100<br/>
            

```python
import numpy as np
```

```python
import keras
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense,Flatten
```

```python
from sklearn.model_selection import train_test_split
```

```python
import matplotlib.pyplot as plt
```

```python
n=100
```

```python
X1=[0]*(n+1)
X2=[0]*(n+1)
X1[0]=1
for i in range(1,len(X1)):
    X1[i]=(X1[i-1])+i*np.sin(i)
for i in range(1,len(X2)):
    X2[i]=(X2[i-1])+i*np.sin(5*i)
    
#X2=X1[::-1]



X=[[[0 for i in range(2)] for j in range(2)]for k in range(n)]
for k in range(len(X)):
    X[k][0][0]=X1[k]
    X[k][0][1]=X2[k]
    
    X[k][1][0]=X1[k+1]
    X[k][1][1]=X2[k+1]

print(X)         
```

```python

Y=[0]*n
for i in range(len(Y)):
    Y[i]=((X[i][0][0]-X[i][0][1])+(X[i][1][0]-X[i][1][1]))
print(Y)
    
```

```python
 for i in range(n):
    print(X[i],Y[i])
```

```python
X=np.array(X, dtype='float32')
Y=np.array(Y,dtype='float32')
```

```python
s=200
X/=s
Y/=s
```

```python
 for i in range(n):
    print(X[i],Y[i])
```

```python
X.shape
```

```python
Y.shape
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2,random_state=5)
```

```python
X_train
```

```python
Y_test.shape
```

```python
from keras.layers import SimpleRNN, LSTM
```

```python
model= Sequential()

model.add(LSTM(2,input_shape=(2,2), return_sequences=True))
#model.add(LSTM(2, return_sequences=False))

#model.add(LSTM((2), return_sequences=False)) # after first layer input_shape_param not req
#model.add(LSTM((3))) ## gives error return_seq paramnecessary
model.add(LSTM((3),return_sequences=True))

model.add(LSTM((4),return_sequences=True))



model.add(Flatten())
model.add(Dense(1,activation="tanh"))

model.compile(optimizer='adam', loss='mse',metrics=['acc'])

model.summary()
```

```python
hist= model.fit(X_train, Y_train, epochs=2000, batch_size=16, validation_data=(X_test, Y_test))
```

```python
Y_predict= model.predict(X_test)
```
