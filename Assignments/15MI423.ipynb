```
import numpy as np
from numpy import array
from keras.models import Sequential
from keras.layers import Dense,Flatten
from keras.layers import SimpleRNN,LSTM
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

X and Y two list(or array are prepared and fed value according to some definite patter)

```
X = list()
Y = list()
X = [x+1 for x in range(30)]
Y = [y * 15 for y in X]

print(X)
print(Y)
```

```
X = array(X).reshape(30, 1, 1) #Reshaping into 3D array
```

```
X.shape
```

```
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(1, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```
model.summary()
```

```
model.fit(X, Y, epochs=1000, validation_split=0.2, batch_size=5)
```

```
test_input = array([30])
test_input = test_input.reshape((1, 1, 1))
test_output = model.predict(test_input, verbose=0)
print(test_output)
```
