```
# Animesh Srivastava 15MI445
import numpy as np
from numpy import array
import math
from keras.models import Sequential
from keras.layers import Dense,Flatten
from keras.layers import SimpleRNN,LSTM
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

In Python, math module contains a number of mathematical operations, which can be performed with ease using the module. math.sin() function returns the sine of value passed as argument. The value passed in this function should be in radians. So, I have modified the function to generate a sequence.

```
X = list()
Y = list()
X = [(math.sin(x * 7) + 5) for x in range(20)]
Y = [y * 3 for y in X]
#Generating sequences
print(X)
print(Y)
```

The output of the above sequence is shown above

```
X = array(X).reshape(20, 1, 1)
# Reshaping the dataset
```

```
X.shape
```

```
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(1, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```
model.summary()
```

```
model.fit(X, Y, epochs=1000, validation_split=0.2, batch_size=5)
```

```
test_input = array([30])
test_input = test_input.reshape((1, 1, 1))
test_output = model.predict(test_input, verbose=0)
print(test_output)
```
